Automating all the process

```python
def text_transformation(X, y, nlp_model, ml_model, coef_show=1):
    
    ## Vectorizing a text data
    X_vectorized = nlp_model.fit_transform(X)
    print("features: {}".format(X_vectorized.shape[1]))
    
    ## Splitting the data into train and test set
    X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y)
    ml = ml_model.fit(X_train, y_train)
    acc = ml.score(X_test, y_test)
    print("Accuracy: ", acc)
    
    if coef_show == 1:
        ## Getting top 20 positive and negative words
        features = nlp_model.get_feature_names()
        coef = ml.coef_.tolist()[0]
        coef_df = pd.DataFrame({"Word": features, "Coefficient":coef})
        coef_df = coef_df.sort_values(["Coefficient", "Word"], ascending=False)
        print("\n")
        print("-Top 20 positive words")
        print(coef_df.head(20))
        print("\n")
        print("-Top 20 negative words")
        print(coef_df.tail(20))
```

```python
text_transformation(X, y, c, LogisticRegression())
```

Automating the prediction 

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
def predict_model(X, y, nlp_model, ml_model):
    ## Vectorizing a text data
    X_vectorized = nlp_model.fit_transform(X)
    
    ## Splitting the data into train and test set
    X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y)
    ml = ml_model.fit(X_train, y_train)
    y_pred = ml.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, cmap="YlGnBu")
    
    acc = accuracy_score(y_test, y_pred)
    print("\n")
    print("The Accuracy is: ", acc)
```

```python
c = CountVectorizer()
```

```python
lr = LogisticRegression()
```

```python
predict_model(X, y, c, lr)
```
