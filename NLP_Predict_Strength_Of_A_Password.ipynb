```python
## Transform the Dataframe into array for easy access for the upcoming operations
password_tuple = np.array(df)
```

```python
password_tuple
```

```python
import random
```

```python
## Let's shuffle our data 
random.shuffle(password_tuple)
```

```python
X = [label[0] for label in password_tuple]
```

```python
y = [label[1] for label in password_tuple]
```

```python
X
```

```python
y
```

```python
## Function to split a password into characters
def divide_word_char(word):
    character = list()
    for i in word:
        character.append(i)
    return character
```

```python
divide_word_char('WUt9IZzE0OQ7PkNE')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=divide_word_char)
```

```python
X_vect = vectorizer.fit_transform(X)
```

```python
X_vect.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_line = X_vect[0]
first_line
```

```python
first_line.T.todense()
```

```python
df2 = pd.DataFrame(first_line.T.todense(), index = vectorizer.get_feature_names(), columns=["TF-IDF"])
```

```python
df2
```

```python
df2.sort_values(by=["TF-IDF"], ascending=False)
```

Let's applying Machine Learning Algorithms

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vect, y, test_size=0.25, random_state=7)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_clf = LogisticRegression(random_state=7, multi_class='multinomial')
```

```python
lg_clf.fit(X_train, y_train)
```

Prediction on a password

```python
ex1 = np.array(["abb@ooo"])
ex1_vect = vectorizer.transform(ex1)
lg_clf.predict(ex1_vect)
```

**GridSearch** to tune the hyperparameters of the `LogisticRgression()`

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = {"C":10.**np.arange(-2,3), "penalty": ["l1", "l2"]}
```

```python
clf_grid = GridSearchCV(estimator = lg_clf, param_grid=grid, cv=5, n_jobs=-1, scoring="f1_macro")
```

```python
clf_grid.fit(X_train, y_train)
```

```python
lg_clf_best = clf_grid.best_estimator_
lg_clf_best
```

**Prediction on the test data**

```python
y_pred = lg_clf_best.predict(X_test)
y_pred
```

**Model evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
